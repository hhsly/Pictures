# An example of ILA-NN implementation


## Load the pre-defined functions, parameters, class. 
If you want to rerun the **code cell**, please run them from top to below since they are connected.    
Double click the cell (or select the cell press `Enter`) and press `control` + `Enter` to run it. Similarly, you can also run this **markdown cell** 

```python
from __future__ import absolute_import, division, print_function, unicode_literals
import tensorflow as tf
from tensorflow.python import keras
import numpy as np
import numpy.fft as npft
import scipy.stats
import scipy.optimize
import scipy.io
import matlab.engine
import matplotlib.pyplot as plt
from matplotlib.mlab import psd
import warnings


class SignalPar(object):
    """
    Some hyper parameters of the PA, saturation, noise power and Volt threshold were measured from weblab
    """
    saturation = 24.1
    impedance = 50   
    noise_power = 0.0032
    pa_gain_in_db = 0
    weblab_lim_in_volt = 0.088

    def __init__(self, sig_length=None, over_rate=None, sig_type='random', fs=200e6):
        """
        Initialize some parameters by default
        
        :param sig_length: The length of the generated signal 
        :param over_rate: Oversampling rate
        :param sig_type: 'Random' signal by default
        :param fs:  Sampling frequency, 200 MHz by default
        """
        self.sig_length = sig_length   # signal length
        self.over_sample_rate = over_rate  # oversampling rate
        self.sig_type = sig_type   
        self.sampling_rate = fs   # sampling frequency
        self.linear_gain = 10 ** (self.pa_gain_in_db / 20)  # Gain in linear scale, only actived for 'tanh' PA
        self.gain_calculated = None   # the calculated normalization gain after receiving PA output
        self.original_sig = None   # original signal before sending to PA
        self.y_desired = None   #  the desired output signal (with gain)
        self.pa_out_no_dpd = None  # the PA output without any DPD 
        
        self.mp_model = 'GMP'  # Define the type of polynomial model; 'GMP' ---> Generalized Memory PolynomialModel
        self.gmp_nonlinear = [6, 6, 6]  # Three nonlinear orders of the GMP model, value > 0
        self.gmp_memory = [6, 6, 6]  # Three memory depths of the GMP model, value > 0
        self.gmp_cross_len = [6, 6]  # Two cross lengths of the GMP model, value, value > 1
        self.papr = None  
        self.rms_in_power = None   # RMS input power
        self.outband_index = None  # out-of-band frequency index
        self.inband_index = None  # in-band frequency index
        
        self.nmse_noDPD = []     # a list to save NMSE result for the case without DPD
        self.nmse_ila_gmp = []   # a list to save NMSE result for ILA with GMP model
        self.nmse_ila_nn = []    # a list to save NMSE result for ILA with neural network
        self.nmse_lb = []        # a list to save NMSE result for lower bound case
        
        self.power_noDPD = []     # a list of the output power for No DPD case
        self.power_ila_gmp = []   # a list of the output power for ILA-GMP case
        self.power_ila_nn = []    # a list of the output power for ILA-NN case
        self.power_lb = []        # a list of the output power for lower bound case
        
        self.psd_noDPD = None    # PSD for No DPD case
        self.psd_ila_gmp = None  # PSD for ILA-GMP case
        self.psd_ila_nn = None   # PSD for ILA-NN case
        self.psd_lb = None       # PSD for Lower Bound case
        
        self.acpr_noDPD = []    # a list of ACPR results for No DPD case
        self.acpr_ila_gmp = []  # a list of ACPR results for ILA-GMP case
        self.acpr_ila_nn = []   # a list of ACPR results for ILA-NN case
        self.acpr_lb = []       # a list of ACPR results for Lower Bound case
        
        self.num_multi_nn = []    # a list of the number of multiplications for ILA-NN case
        self.num_multi_gmp = []   # a list of the number of multiplications for ILA-GMP case
        self.num_par_nn = []    # a list of the number of parameters for ILA-NN case
        self.num_par_gmp = []   # a list of the number of parameters for ILA-GMP case

    def signal_generator(self, added_power_in_dbm):
        """
        Generate the OFDM signal by specifying the added power, signal length, oversampling rate
        
        :param added_power_in_dbm:  A given added power in dBm
        """
        if self.sig_type == 'random':
            sig = (1 / np.sqrt(2)) * (np.random.randn(self.sig_length, 1) + 1j * np.random.randn(self.sig_length, 1))
            sig = self.ideal_filter(sig, 1/self.over_sample_rate)
            sig = sig / np.linalg.norm(sig, 2)
            self.papr = 20*np.log10(max(abs(sig)) * np.sqrt(self.sig_length)/np.linalg.norm(sig, 2))
            self.original_sig = 10**(added_power_in_dbm/20) * sig
            self.rms_in_power = 10*np.log10(np.linalg.norm(self.original_sig, 2)**2/(2*self.impedance)/self.sig_length) + 30

    def ideal_filter(self, signal, band_rate):
        """
        Given the time-domain signal 'signal', filter out the unwanted frequencies by specifying the oversampling rate
        
        :param signal: Time-domain signal
        :param band_rate:  1/oversampling rate
        :return: An OFDM-like signal
        """
        # FFT the input
        sig_fft = npft.fft(signal, axis=0)
        # compute the needed index
        N = int(np.ceil(0.5 * self.sig_length*band_rate))
        # Null the out-of-band frequencies bins
        sig_fft[N:-N] = 0
        # Remove the DC
        sig_fft[0] = 0
        # save the inband index
        self.outband_index = list(range(N, 3*N))
        self.inband_index = list(range(0, N)) + list(range(self.sig_length-N, self.sig_length))
        return npft.ifft(sig_fft, axis=0)

    def call_power_amplifier(self, pa_in, pa_model, eng=None):
        """
        Call the power amplifier model
        
        :param pa_in: Input of the PA
        :param pa_model: PA model to be called
        :return: PA output
        """
        if pa_model == 'tanh':
            sat = self.saturation
            noise = np.sqrt(self.noise_power/2) * (np.random.randn(self.sig_length, 1) + 1j*np.random.randn(self.sig_length, 1))
            return self.linear_gain*(sat*np.tanh(abs(pa_in)/sat) * np.exp(1j*np.angle(pa_in)) + noise)
        if pa_model == 'weblab':
            check_out = self.check_inputs_weblab(pa_in)
            # Transfer the numpy.array to Matlab data type
            scipy.io.savemat('check_out.mat', {'check_out': check_out})
            check_out_matlab = eng.load('check_out.mat', nargout=1)
            print('Waiting for measurement from WebLab...')
            pa_out_matlab_time_align = eng.call_weblab_and_timealign(check_out_matlab['check_out'])
            print('Results are ready')
            return pa_out_matlab_time_align
        if pa_model == 'ideal':
            sat = self.saturation
            noise = np.sqrt(self.noise_power/2) * (np.random.randn(self.sig_length, 1) + 1j*np.random.randn(self.sig_length, 1))
            pa_in[abs(pa_in) > sat] = sat*np.exp(1j*np.angle(pa_in[abs(pa_in) > sat]))
            return self.linear_gain*(pa_in + noise)

    def check_inputs_weblab(self, x):
        """
        Check the PAPR, RMS power and other requirements of the signal 'x' before send to RF WebLab due to the rule of WebLab
        Should the above requirements are not satisfied, the signal be clipped by a specific amplitude threshold.
        :param x:  Signal that need to be checked
        :return:  signal after the check
        """
        rms_power_in = 10*np.log10(np.linalg.norm(x, 2) ** 2 / self.impedance / self.sig_length) + 30
        papr = 20*np.log10(max(abs(x)) * np.sqrt(self.sig_length)/np.linalg.norm(x, 2))
        threshold_volt = self.weblab_lim_in_volt

        if rms_power_in > -8 - papr*0.77:
            warnings.warn('RF weblab input check failed: rms input power too high', DeprecationWarning)
            # Cut the peaks of x so as to meet the requirement
            x[abs(x) > threshold_volt] = threshold_volt * x[abs(x) > threshold_volt] / abs(x[abs(x) > threshold_volt])
            
        if papr > 20:
            warnings.warn('RF weblab input check failed: Signal PAPR too high', DeprecationWarning)
            # Find the smaller threshold between volt threshold and PAPR threshold
            papr_threshold = min(threshold_volt, np.sqrt(10 ** (20 / 10) * np.var(x)))
            # Cut the peaks of x so as to meet the requirement
            x[abs(x) > papr_threshold] = papr_threshold * x[abs(x) > papr_threshold] / abs(x[abs(x) > papr_threshold])
            rms_power_in = 10 * np.log10(np.linalg.norm(x, 2) ** 2 / self.impedance / self.sig_length) + 30
            papr = 20 * np.log10(max(abs(x)) * np.sqrt(self.sig_length) / np.linalg.norm(x, 2))

        if rms_power_in + papr > -8:
            warnings.warn('RF weblab input check failed: Peak input power too high', DeprecationWarning)
            # Cut the peaks of x so as to meet the requirement
            x[abs(x) > threshold_volt] = threshold_volt * x[abs(x) > threshold_volt] / abs(x[abs(x) > threshold_volt])
        return x

    def dpd_methods(self, pa_in=None, pa_out=None, method=None, pa_model=None, psd_plot=False, matlab_eng=None):
        """
        A function operates different DPD methods, calculate NMSE and ACPR, and may plot the PSD
        
        :param pa_in:  Input signal of the PA
        :param pa_out:  Output signal of the PA
        :param method:  DPD method, e.g., 'ILA-NN', 'No DPD', 'ILA-GMP', 'Lower Bound'
        :param pa_model:  Type of PA model, 'Weblab' or 'tanh'
        :param psd_plot:  A flag; True -> Plot PSD, False -> Not plot PSD
        """
        if method == 'ILA-NN':
            # Transfer a two-column real value array  to a one/column complex value array
            pa_in_complex = pa_in[:, 0:1] + 1j*pa_in[:, 1:]
            pa_out = self.call_power_amplifier(pa_in=pa_in_complex, pa_model=pa_model, eng=matlab_eng)
            if pa_model == 'tanh':
                # Calculate output power
                self.power_ila_nn.append(10*np.log10(np.linalg.norm(pa_out / self.gain_calculated, 2)**2 /
                                                     (2*self.impedance) / self.sig_length) + 30)
            else:
                self.power_ila_nn.append(10 * np.log10(np.linalg.norm(pa_out, 2) ** 2 /
                                                       (2 * self.impedance) / self.sig_length) + 30)
            # Calculate NMSE
            self.nmse_ila_nn.append(10*np.log10(np.var(self.y_desired - pa_out)/np.var(self.y_desired)))
            # Plot the PSD of the PA output
            self.plot_psd(pa_out, name=method, num_fig=5, plot_flag=psd_plot, plot_ending='psd')
            # Plot the error spectra between the PA output and the desired output
            self.plot_psd(pa_out - self.y_desired, name=method, num_fig=6, plot_flag=psd_plot, plot_ending='psd-error')
            # Calculate PSD
            self.psd_ila_nn = abs(npft.fft(pa_out, axis=0))**2/len(pa_out)/self.sampling_rate
            # Calculate ACPR
            self.acpr_ila_nn.append(10*np.log10(np.sum(self.psd_ila_nn[self.outband_index])
                                                / np.sum(self.psd_ila_nn[self.inband_index])))
            print('Current NMSEs ==> ILA-NN: ', self.nmse_ila_nn[-1], ' dBm')

        elif method == 'Lower Bound':
            # Normalize the added Gain in tanh PA model
            if pa_model == 'tanh':
                pa_out = pa_out / self.gain_calculated
            # Calculate the PA's output signal variance
            pa_out_var = np.var(pa_out)/2
            sat = self.saturation
            # Theoretical power equation
            self.power_lb.append(10*np.log10(1e3 / self.impedance *
                                             (pa_out_var - pa_out_var * np.exp((-sat**2)/(2*pa_out_var)) +
                                              self.noise_power/2)))
            # Theoretical NMSE equation,  scipy.stats.norm.sf() is the Q-function
            self.nmse_lb.append(10*np.log10((2*pa_out_var * np.exp(-sat**2/(2*pa_out_var)) -
                                             2*sat*np.sqrt(pa_out_var*2*np.pi) *
                                             scipy.stats.norm.sf(sat/np.sqrt(pa_out_var)) + self.noise_power) /
                                            (2*pa_out_var)))
            print('Current NMSEs ==> Lower Bound: ', self.nmse_lb[-1], ' dBm\n')
            # Plot the desired PSD
            # self.plot_psd(self.y_desired, name='Desired output', num_fig=5, plot_flag=psd_plot, plot_ending=True)
            
            # Pass the desired signal to an ideal PA
            pa_out_ideal = self.call_power_amplifier(pa_in=self.y_desired, pa_model='ideal')
            # Calculate PSD of the ideal PA output as lower bound simulations
            self.psd_lb = abs(npft.fft(pa_out_ideal, axis=0))**2/len(pa_out_ideal)/self.sampling_rate
            # Calculate ACPR of the ideal PA output as lower bound simulations
            self.acpr_lb.append(10*np.log10(np.sum(self.psd_lb[self.outband_index])
                                                / np.sum(self.psd_lb[self.inband_index])))


        elif method == 'No DPD':
            if pa_model == 'tanh':
                # Calculate output power
                self.power_noDPD.append(10*np.log10(np.linalg.norm(pa_out / self.gain_calculated, 2)**2 /
                                                    (2*self.impedance) / self.sig_length) + 30)
            else:
                # Calculate output power
                self.power_noDPD.append(10 * np.log10(np.linalg.norm(pa_out, 2) ** 2 /
                                                      (2*self.impedance) / self.sig_length) + 30)
            # Calculate NMSE
            self.nmse_noDPD.append(10 * np.log10(np.var(self.y_desired - pa_out) / np.var(self.y_desired)))
            # Plot the PSD of PA output
            self.plot_psd(pa_out, name=method, num_fig=5, plot_flag=psd_plot)
            # Plot the error spectra between the PA output and the desired output
            self.plot_psd(pa_out-self.y_desired, name=method, num_fig=6, plot_flag=psd_plot)
            # Calculate PSD
            self.psd_noDPD = abs(npft.fft(pa_out, axis=0)) ** 2/len(pa_out)/self.sampling_rate
            # Calculate ACPR
            self.acpr_noDPD.append(10 * np.log10(np.sum(self.psd_noDPD[self.outband_index]) /
                                                 np.sum(self.psd_noDPD[self.inband_index])))
            print('Current Output RMS power: ', self.power_noDPD[-1], ' dBm')
            print('Current NMSEs ==> No DPD: ', self.nmse_noDPD[-1], ' dBm')

        elif method == 'ILA-GMP':
            # Get the matrix version of PA output
            pa_out_matrix = self.memory_polynomial_model(pa_out)
            # Solve the post-distorter parameters by Least-Squares method
            coe_ila_dpd, residuals = np.linalg.lstsq(pa_out_matrix, pa_in, rcond=-1)[0:2]
            # Using the post-distorter parameters as the pre-distorter parameters, and calculate the DPD output
            dpd_in_matrix = self.memory_polynomial_model(pa_in)
            dpd_output = np.dot(dpd_in_matrix, coe_ila_dpd)
            # DPD output as the PA input
            pa_out_with_dpd = self.call_power_amplifier(pa_in=dpd_output, pa_model=pa_model, eng=matlab_eng)
            if pa_model == 'tanh':
                # Calculate output power
                self.power_ila_gmp.append(10*np.log10(np.linalg.norm(pa_out_with_dpd / self.gain_calculated, 2)**2 /
                                                      (2*self.impedance) / self.sig_length) + 30)
            else:
                # Calculate output power
                self.power_ila_gmp.append(10 * np.log10(np.linalg.norm(pa_out_with_dpd, 2) ** 2 /
                                                        (2*self.impedance) / self.sig_length) + 30)
            # Plot the PSD
            self.plot_psd(pa_out_with_dpd, name=method, num_fig=5, plot_flag=psd_plot)
            # Plot the error spectra between the PA output and the desired output
            self.plot_psd(pa_out_with_dpd - self.y_desired, name=method, num_fig=6, plot_flag=psd_plot)
            # Calculate NMSE
            self.nmse_ila_gmp.append(10 * np.log10(np.var(self.y_desired - pa_out_with_dpd) / np.var(self.y_desired)))
            # Calculate PSD
            self.psd_ila_gmp = abs(npft.fft(pa_out_with_dpd, axis=0)) ** 2/len(pa_out)/self.sampling_rate
            # Calculate ACPR
            self.acpr_ila_gmp.append(10 * np.log10(np.sum(self.psd_ila_gmp[self.outband_index]) /
                                                   np.sum(self.psd_ila_gmp[self.inband_index])))
            print('Current NMSE ==> ILA-GMP: ', self.nmse_ila_gmp[-1], ' dBm, and the LS residuals are ', residuals)

    def memory_polynomial_model(self, x):
        """
        Build the memory polynomial model matrix, form the matrix X in the equation: Y = X*w.
        Details can be found in the paper "A GeneralizedMemory PolynomialModelfor Digital Predistortion of RF Power Ampliï¬ers"
        :param x: input signal
        :return: x_matrix, the formed input matrix
        """
        # nonlinear orders, K[0], K[1] and K[2] are the nonlinear order of MP envelope,
        # lagging envelope and leading envelope of GMP model
        K = self.gmp_nonlinear
        M = self.gmp_memory  # Memory lengths of the GMP; M[0], M[1], M[2]
        N = self.gmp_cross_len  # Cross lengths of the GMP; N[0], N[1]
        x_len = x.shape[0]
        # First, form the MP envelope of GMP, number of parameters: K[0]*M[0]
        if M[0] > 0:
            x_memory = np.vstack((x[-M[0]:, :], x))
        else:
            x_memory = x
        for k in range(0, K[0]):
            for m in range(0, M[0]):
                if m == 0:
                    x_array = abs(x_memory[M[0]:, :]) ** k * x_memory[M[0]:, :]
                else:
                    x_array = np.hstack((x_array, abs(x_memory[M[0] - m: -m, :]) ** k * x_memory[M[0] - m: -m, :]))
            if k == 0:
                x_matrix = x_array
            else:
                x_matrix = np.hstack((x_matrix, x_array))
        # Then, form the lagging envelope of GMP, number of parameters: K[1]*M[1]*L[0]
        if self.mp_model == 'GMP':
            x_memory = np.vstack((x[-M[1]-N[1]:, :], x))
            for k in range(1, K[1]+1):
                for m in range(0, M[1]):
                    for n in range(1, N[0]+1):
                        idx = M[1] + N[0] - n - m
                        if n == 1:
                            x_array_cross = abs(x_memory[idx: idx+x_len, :]) ** k * x_memory[idx+n: idx+n+x_len, :]
                        else:
                            x_array_cross = np.hstack((x_array_cross, abs(x_memory[idx: idx+x_len, :]) ** k *
                                                       x_memory[idx+n: idx+n+x_len, :]))
                    if m == 0:
                        x_array = x_array_cross
                    else:
                        x_array = np.hstack((x_array, x_array_cross))
                x_matrix = np.hstack((x_matrix, x_array))

        # Finally, form the leading envelope of GMP, number of parameters: K[2]*M[2]*L[1]
            x_memory = np.vstack((x[-M[2]:, :], x))
            x_memory = np.vstack((x_memory, x[:N[1]+1, :]))
            for k in range(1, K[2]+1):
                for m in range(0, M[2]):
                    for n in range(1, N[1]+1):
                        idx = M[2] + n - m
                        if n == 1:
                            x_array_cross = abs(x_memory[idx: idx+x_len, :]) ** k * x_memory[idx-n: idx-n+x_len, :]
                        else:
                            x_array_cross = np.hstack((x_array_cross, abs(x_memory[idx: idx+x_len, :]) ** k *
                                                       x_memory[idx-n: idx-n+x_len, :]))
                    if m == 0:
                        x_array = x_array_cross
                    else:
                        x_array = np.hstack((x_array, x_array_cross))
                x_matrix = np.hstack((x_matrix, x_array))
        return x_matrix

    def plot_nmse_and_acpr(self, num_figure):
        """
        A function to plot the NMSEs and ACPR results
        
        :param num_figure: The Figure index which the results are plotted on
        :return: Two Figures
        """
        
        plt.figure(num_figure)
        plt.plot(np.array(self.power_noDPD), np.array(self.nmse_noDPD), '-s',
                 linewidth=2, markersize=12, fillstyle='none', label='Without DPD')
        plt.plot(np.array(self.power_ila_gmp), np.array(self.nmse_ila_gmp), '-o',
                 linewidth=2, markersize=12, fillstyle='none', label='ILA-GMP')
        plt.plot(np.array(self.power_ila_nn), np.array(self.nmse_ila_nn), '-p',
                 linewidth=2, markersize=12, fillstyle='none', label='ILA-NN')
        plt.plot(np.array(self.power_lb), np.array(self.nmse_lb), '-*',
                 linewidth=3, markersize=12, fillstyle='none', label='Lower Bound')
        plt.xlabel('Avg. output power (dBm)')
        plt.ylabel('NMSE (dB)')
        plt.legend(loc=0)
        plt.grid()
        plt.show()

        plt.figure(num_figure+1)
        plt.plot(np.array(self.power_noDPD), np.array(self.acpr_noDPD), '-s',
                 linewidth=2, markersize=12, fillstyle='none', label='Without DPD')
        plt.plot(np.array(self.power_ila_gmp), np.array(self.acpr_ila_gmp), '-o',
                 linewidth=2, markersize=12, fillstyle='none', label='ILA-GMP')
        plt.plot(np.array(self.power_ila_nn), np.array(self.acpr_ila_nn), '-p',
                 linewidth=2, markersize=12, fillstyle='none', label='ILA-NN')
        plt.plot(np.array(self.power_lb), np.array(self.acpr_lb), '-*',
                 linewidth=3, markersize=12, fillstyle='none', label='LB simulation')
        plt.xlabel('Avg. output power (dBm)')
        plt.ylabel('ACPR (dB)')
        plt.legend(loc=0)
        plt.grid()
        plt.show()
       

    def plot_psd(self, x, name, num_fig, plot_flag=False, plot_ending=None):
        """
        A function to plot the PSD of a given signal
        
        :param x: The input signal
        :param name: The name of the plotted line
        :param num_fig: Figure index
        :param plot_flag: A Flag; True -> Plot the figure, False -> Not plot the figure
        :param plot_ending: A Flag; 'psd' -> The PSD is plotted,  'psd-error' -> The PSD error is plotted
        :return: A Figure
        """
        if plot_flag:
            plt.figure(num_fig)
            x_array = np.asarray(x)
            pxx, freqs = psd(x_array[:, 0], NFFT=512, Fs=200e6, scale_by_freq=False)
            plt.plot(freqs*1e-6, 10*np.log10(pxx), label=name, linewidth=2)
            plt.legend(loc=0)
            # plt.plot(10*np.log10(abs(npft.fftshift(npft.fft(x_array, n=2e5, axis=0)))**2/200e6/2e5))
            
            if plot_ending == 'psd':
                plt.figure(num_fig)
                plt.xlabel('Frequency [MHz]')
                plt.ylabel('Power Spectral Density (dB)')
                plt.xlim(-100, 100)
                plt.grid()
                plt.legend(loc=0)
            if plot_ending == 'psd-error':
                plt.figure(num_fig)
                plt.xlabel('Frequency [MHz]')
                plt.ylabel('PSD of error (dBc)')
                plt.xlim(-100, 100)
                plt.grid()
                plt.legend(loc=0)


def slice_train_test_data(data_in, data_out, mem_len=0):
    """
    Separate the complex-valued input/output to real-valued signal,
    form the input/output by absconsidering the memory effects,
    normalize the data into [-1,1].
    
    :param data_in:  Input data of the neural network
    :param data_out:  Output data of the neural network
    :param mem_len:  Memory length of the Input 
    :return:
    """
    x_total = np.hstack((data_in.real, data_in.imag))
    y_total = np.hstack((data_out.real, data_out.imag))
    
    # Scale the data into [-1,1]
    scale = x_total.max()
    x_total = x_total/scale
    y_total = y_total/scale
    
    # Form the input matrix by specifying the memory length
    x_memory = stack_memory(x_total, mem_len=mem_len)
    y_memory = stack_memory(y_total, mem_len=mem_len)

    return x_total, y_total, x_memory, y_memory, scale


def stack_memory(input, mem_len=0):
    """
    Form the input of the neural network with memory;
    E.g., input is a 20000x2 matrix, mem_len=3 ==> output is a 20000x8 matrix
    :param input: The current instance input matrix, e.g., 200000x2
    :param mem_len: memory depth of the NN input layer
    :return: NN input with memory
    """
    dim = input.shape[1]
    for i in range(0, mem_len):
        input_mem = np.vstack((input[-i-1:, 0:dim], input[0:-i-1, 0:dim]))
        input = np.hstack((input, input_mem))
    return input


def plot_history(histories, current_factor, num_figure, key='mse', ):
    """
    Plot the training history (e.g., MSE loss) against the number of training epoch
    
    :param histories:  Model.history; The saved training history which contains
    :param current_factor:  Index of the current Power 
    :param num_figure:  Figure index
    :param key:  The name of the loss function results in the histories
    """
    for name, history in histories:
        plt.figure(num_figure)
        val = plt.plot(history.epoch, history.history['val_' + key],
                       '--', label=name.title() + ' Val' + str(current_factor))
        plt.plot(history.epoch, history.history[key], color=val[0].get_color(),
                 label=name.title() + ' Train ' + str(current_factor))
    plt.xlabel('Epochs')
    plt.ylabel(key.replace('_', ' ').title())
    plt.legend()
    plt.xlim([0, max(history.epoch)])


def build_model(num_neuron, name):
    """
        given parameters, build a keras.model
    :param num_neuron: list, the number of neurons for each layer
    :param name: str, the name of the model
    :return: keras.Model object
    """

    input_layer = keras.Input(shape=(num_neuron[0],))  # create the input layer
    x = keras.layers.Dense(units=num_neuron[1])(input_layer)  # add the second layer
    x = keras.layers.ReLU(max_value=None, negative_slope=0)(x)
    for num in num_neuron[2:-1]:
        x = keras.layers.Dense(units=num)(x)
        x = keras.layers.ReLU(max_value=None, negative_slope=0.1)(x)
    x = keras.layers.Dense(units=num_neuron[-1])(x)
    # output = keras.layers.add([input_layer, x])  # Add the input to the output directly
    return keras.Model(inputs=input_layer, outputs=x, name=name)

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "last"
%config InlineBackend.figure_format = 'svg'   # Change the figure format to .svg instead of .png
```

## Build the neural network
Let's define a simple neural network with only one hide layer.   
First, we separate the complex-valued signal to real and imaginary parts. We consider the input with a memory depth `nn_mem_len`, so the input layer has `2(nn_mem_len +1)` neurons. Consider `512` neurons in the hide layer and `2` neurons in the output layer.   
![NN structure](https://github.com/hhsly/Pictures/raw/master/ILA-NN.png)

```python
# Create the input layer with shape 2*(nn_mem_len+1)
input_layer = keras.Input(shape=(2*(nn_mem_len+1),))

# Add the second layer which is a fully connected layer (Dense) with 512 neurons
x = keras.layers.Dense(units=512)(input_layer)

# The activation function of the second layer is a standard RuLU
x = keras.layers.ReLU(max_value=None, negative_slope=0)(x)

# The output layer has two neurons for the real and imaginary signal.
output = keras.layers.Dense(units=2)(x)  

# Using keras.Model to build this network by specifying the input and output, also give it a name 'Post-distorter'
model_dpd_post = keras.Model(inputs=input_layer, outputs=output, name='Post-distorter')

# Compile the model by specifying the optimizer, loss function, metric. Here we use Adam, MSE and MSE
model_dpd_post.compile(optimizer='adam', loss='mse', metrics=['mse'])

# Let's check the structure of 'model_dpd_post'
model_dpd_post.summary()

# Plot the diagram of 'model_dpd_post'
keras.utils.plot_model(model_dpd_post, show_shapes=True)
```

## Generate the input signal with a given power level

```python
# Define the added power in dBm
power_add = 16 

# Generate a Gaussian complex-valued signal with N(0,1)
sig = (1 / np.sqrt(2)) * (np.random.randn(par.sig_length, 1) + 1j * np.random.randn(par.sig_length, 1))
print('Variance of sig:', np.var(sig))

# Get the desired oversampled signal by filtering, the output 'sig_filter' is an OFDM-liked signal
sig_filter = par.ideal_filter(sig, 1/par.over_sample_rate)
# Normalize the power of the signal
sig_filter = sig_filter / np.linalg.norm(sig_filter, 2)

# Calculate the initial PAPR
par.papr = 20*np.log10(max(abs(sig_filter)) * np.sqrt(par.sig_length)/np.linalg.norm(sig_filter, 2))

# Add the specified power to the signal 
par.original_sig = 10**(power_add/20) * sig_filter

# Calculate the RMS power after power addition step
par.rms_in_power = 10*np.log10(np.linalg.norm(par.original_sig, 2)**2/(2*par.impedance)/par.sig_length) + 30

print('Current added power is: ', power_add, ' dBm')
print('Current Input RMS power is: ', par.rms_in_power, 'dBm')

# Plot the PSD of the generated signal
plt.psd(par.original_sig[:,0], NFFT=1024, Fs=200e6)
# plt.plot(10*np.log10(abs(npft.fftshift((npft.fft(par.original_sig, axis=0))))**2/200e6/2e5))
```

## Form the training data and train the ILA-based NN

1. Function `slice_train_test_data()` separates the complex-valued input/output to real-valued signal, forms the input/output considering the memory depth `nn_men_len`, also normalizes the data into [-1,1]
2. Then, train the pre-defined NN model `model_dpd_post`. According to the definition of ILA, the NN's **input** `y_memory` is formed by PA's **output** `par.y_power_align`, and the NN's **output** `x_total` is formed by the PA's **input** `par.original_sig` 

```python
x_total, y_total, x_memory, y_memory, normalize_scale = slice_train_test_data(
    data_in=par.original_sig, data_out=par.y_power_align, mem_len=nn_mem_len)

print('\nTraining the PA inverse model (post-distorter)...')
# Train the post-predistorter whose input is the PA output, save the training history to 'model_history_post_dpd'
model_history_post_dpd = model_dpd_post.fit(y_memory, x_total, batch_size=128, epochs=num_epochs,
                                                verbose=1, validation_split=0.1)
```

## After training, use the NN model to generate the pre-distorted signal

In ILA, the pre-distorter is a copy of the post-distorter `model_dpd_post`, so after training, we can consider the `model_dpd_post` as the pre-distorter and predict the distorted signal with the input `x_memory`, which is formed by the PA's **input** `par.original_sig` with memory depth `nn_mem_len`

```python
# After training, predict the NN output
x_predict_post = model_dpd_post.predict(x_memory)

# Since we normalized the PA output power in 'slice_train_test_data()' for proper training results, here should be recovered  
pa_in = x_predict_post * normalize_scale

# Transfer a two-column real value array  to a one/column complex value array
pa_in_complex = pa_in[:, 0:1] + 1j*pa_in[:, 1:]

# Call the PA model
pa_out = par.call_power_amplifier(pa_in=pa_in_complex, pa_model=pa_model, eng=eng)
```

## Calculate the results (NMSE, PSD, PSD-error, ACPR) at the given output power

Once we received PA output, we can calculate the output power `power_ila_nn`, the NMSE between PA output and the desired output (`nmse_ila_nn`), the PSD of PA output, the PSD error between PA output and the desired output, and the ACPR results (`acpr_ila_nn`)

```python
# Calculate output power of ILA-NN
par.power_ila_nn.append(10 * np.log10(np.linalg.norm(pa_out, 2) ** 2 /
                                           (2 * par.impedance) / par.sig_length) + 30)
# Calculate NMSE of ILA-NN
par.nmse_ila_nn.append(10*np.log10(np.var(par.y_desired - pa_out)/np.var(par.y_desired)))

# Plot the PSD of the PA output
par.plot_psd(pa_out, name='ILA-NN', num_fig=5, plot_flag=True, plot_ending='psd')

# Plot the error spectra between the PA output and the desired output
par.plot_psd(pa_out - par.y_desired, name='ILA-NN', num_fig=6, plot_flag=True, plot_ending='psd-error')

# Calculate PSD
par.psd_ila_nn = abs(npft.fft(pa_out, axis=0))**2/len(pa_out)/par.sampling_rate

# Calculate ACPR
par.acpr_ila_nn.append(10*np.log10(np.sum(par.psd_ila_nn[par.outband_index])
                                    / np.sum(par.psd_ila_nn[par.inband_index])))

print('Current NMSEs ==> ILA-NN: ', par.nmse_ila_nn[-1], ' dBm')
print('Current ACPRs ==> ILA-NN: ', par.acpr_ila_nn[-1], ' dBm')
```

# An example to generate and show results with a range of output power

In the previous example we present the steps of `ILA-NN` implementation with a specific added power `power_add`. In this section we will evaluate the performance over **a range of output power** by giving `power_add_new` a range.

- **Notice:** rerun this will take around **40 minutes** to get the final results because of the loop.

- Because of the highly nonlinear effect of PA at a high output power, the performance of `ILA-GMP` suddenly becomes bad at the last output power level (30.8 dBm). See the last point of NMSE and ACPR figures. In addition, the least-squares estimation of the high order terms coefficients in GMP model has a large error, which makes the high-order terms dominate the error. At this power level, lower the nonlinear order `gmp_nolinear` will relief the problem and has a better performance.

```python
# Added power from 0 to 22 with a step size 2. E.g., 0, 2, 4, ... ,22. 
power_add_new = range(0, 23, 2)

# Define an new object 'par_new'
par_new = SignalPar(sig_length=int(2e5), over_rate=10) 
pa_model = 'weblab'  
num_epochs = 10  # number of epochs per training process
nn_mem_len = 7  # Memory depth of the NN input layer
par_new.gmp_nonlinear = [5, 5, 5]  # Three nonlinear orders of the GMP model, value > 0
par_new.gmp_memory = [6, 6, 6]  # Three memory depths of the GMP model, value > 0
par_new.gmp_cross_len = [6, 6]  # Two cross lengths of the GMP model, value, value > 1

model_dpd_post_new = build_model([2*(nn_mem_len+1), 640, 2], name='post_distorter')
model_dpd_post_new.compile(optimizer='adam', loss='mse', metrics=['mse'])

# Start the MATLAB engine
eng_new = matlab.engine.start_matlab()

for i in power_add_new:
    par_new.signal_generator(i)
    par_new.pa_out_no_dpd = par_new.call_power_amplifier(pa_in=par_new.original_sig, pa_model=pa_model, eng=eng_new)
    par_new.gain_calculated = np.linalg.norm(par_new.pa_out_no_dpd, 2)/np.linalg.norm(par_new.original_sig, 2)
    
    # Compute the desired output
    par_new.y_desired = par_new.original_sig * par_new.gain_calculated
    # Normalize the output signal power level to the input signal power level
    par_new.y_power_align = par_new.pa_out_no_dpd / par_new.gain_calculated
    
    x_total, y_total, x_memory, y_memory, normalize_scale = slice_train_test_data(
            data_in=par_new.original_sig, data_out=par_new.y_power_align, mem_len=nn_mem_len)
    print('\nTraining the PA inverse model (post-distorter)...')
    
    model_history_post_dpd_new = model_dpd_post_new.fit(y_memory, x_total, batch_size=128, epochs=num_epochs,
                                                    verbose=0, validation_split=0.1)
    x_predict_post = model_dpd_post_new.predict(x_memory)
    print('Current PA model is: ', pa_model)
    print('Current added power is: ', i, ' dBm')
    print('Current Input RMS power: ', par_new.rms_in_power, ' dBm')
    
    par_new.dpd_methods(pa_out=par_new.pa_out_no_dpd, method='No DPD', pa_model=pa_model, psd_plot=False)
    
    # ILA method using GMP model, the PA output is power aligned
    par_new.dpd_methods(pa_in=par_new.original_sig, pa_out=par_new.y_power_align, method='ILA-GMP', 
                        pa_model=pa_model, psd_plot=False, matlab_eng=eng_new)
   
    # ILA method using neural networks, the NN output should be rescaled by the same scale as
    # in the Function 'slice_train_test_data()'
    par_new.dpd_methods(pa_in=x_predict_post * normalize_scale, method='ILA-NN', pa_model=pa_model, psd_plot=False,
                        matlab_eng=eng_new)

    # Lower bound calculation
    par_new.dpd_methods(pa_out=par_new.pa_out_no_dpd, method='Lower Bound', pa_model=pa_model)
    
    # Plot the training history (loss versus epoch at each power level)
    plot_history([('Post-distorter', model_history_post_dpd_new)], i - power_add_new[0], num_figure=1)

par_new.plot_nmse_and_acpr(num_figure=3)
```

# An example of complexity comparison between `ILA-NN` and `ILA-GMP`

In this example, we would compare the computation complexity of `ILA-NN` and `ILA-GMP`.   
**Notice:** Rerun this example will take around **1.5 hours**.

## The codes: 
 The steps are shown below:
1. Iterate the number of memory depth in both `ILA-NN` and `ILA-GMP`
2. For `ILA-NN`: At each memory depth, iterate the number of neurons in the hide layer (only one hide layer is considered here).
3. For `ILA-GMP`: At each memory depth, iterate the number of non-linear order (cross length is equal to the memory length here).
4. Plot the scatter points of `ILA-NN` and `ILA-GMP` at each memory iteration.

```python
from itertools import cycle
import numpy as np
import matplotlib.pyplot as plt


par_multi = SignalPar(sig_length=int(2e5), over_rate=10)  # Define an object 'par' which contains a lot of parameters
pa_model = 'weblab'  # Define the type of PA model that are used: 'tanh', 'weblab'
train_prop = 1  # the proportion of training data in all data; e.g. 0.8 -> 80% training data/ 20% test data
num_epochs = 10  # Number of epochs
neur_range = [8, 16, 32, 64, 128, 256, 512, 1024, 2048]   # Number of neurons in the hide layer
nn_mem_len = 7  # Largest memory depth of the NN input layer
nonlinear_order = 7  # Largest nonlinear order

# Start the MATLAB engine
eng_multi = matlab.engine.start_matlab()

# added power to the input signal, dBm, the generated initial signal power is -43 dBm
power_add = 16  # Set the input power from -24 dBm to -8 dBm for 'weblab' PA
par_multi.signal_generator(power_add)  # Generate signal with specific added power

# Get PA output
par_multi.pa_out_no_dpd = par_multi.call_power_amplifier(pa_in=par_multi.original_sig, pa_model=pa_model, eng=eng_multi)  
# Calculate the gain of the PA
par_multi.gain_calculated = np.linalg.norm(par_multi.pa_out_no_dpd, 2)/np.linalg.norm(par_multi.original_sig, 2)
# Compute the desired output
par_multi.y_desired = par_multi.original_sig * par_multi.gain_calculated
# Normalize the output signal power level to the input signal power level
par_multi.y_power_align = par_multi.pa_out_no_dpd / par_multi.gain_calculated
# Color map for plotting
color_map = cycle('bgrcmykw')

# ILA-NN:
# Iterate the input layer memory depth from 0 to nn_mem_len-1 of ILA-NN
for i in range(0, nn_mem_len):
    print('Current memory depth: ', i)
    # Iterate the number of neurons in the range 'neur_range'
    for neuron in neur_range:
        print('\nCurrent number of neurons in the hide layer: ', neuron)
        plot_length = len(neur_range)
        model_dpd_post = build_model([2 * (i + 1), neuron, 2], name='post_distorter')
        model_dpd_post.compile(optimizer='adam', loss='mse', metrics=['mse'])
        par_multi.num_multi_nn.append((2 * i + 4) * neuron)
        par_multi.num_par_nn.append((2*i+5)*neuron+2)
        print('Neural Network Structure: ', 2 * (i + 1), '-', neuron, '-2')
        print('Number of parameters: ', par_multi.num_par_nn[-1])
        print('Number of multiplications: ', par_multi.num_multi_nn[-1])
        x_total, y_total, x_memory, y_memory, normalize_scale = slice_train_test_data(
            data_in=par_multi.original_sig, data_out=par_multi.y_power_align, mem_len=i)
        
        # Train the NN model
        model_history_post_dpd = model_dpd_post.fit(y_memory, x_total, batch_size=128, epochs=num_epochs,
                                                    verbose=0, validation_split=0.1)
        
        # Predict the pre-distorted signal using the trained NN model
        x_predict_post = model_dpd_post.predict(x_memory)
        # Pass the pre-distorted signal to PA and calculate results
        par_multi.dpd_methods(pa_in=x_predict_post * normalize_scale, method='ILA-NN', pa_model=pa_model, matlab_eng=eng_multi)

    color = next(color_map)
    plt.figure(2)
    # Plot the NMSE results of ILA-NN at current input memory depth
    plt.scatter(par_multi.num_multi_nn[-plot_length:], par_multi.nmse_ila_nn[-plot_length:], marker='o', c=color, label='M='+str(i)+', ILA-NN')
    plt.figure(3)
    # Plot the ACPR results of ILA-GMP at current input memory depth
    plt.scatter(par_multi.num_multi_nn[-plot_length:], par_multi.acpr_ila_nn[-plot_length:], marker='o', c=color, label='M='+str(i)+', ILA-NN')

    # Calculate the results for ILA-GMP
    # Iterate the nonlinear order from 1 to nonlinear_order-1
    for k in range(1, nonlinear_order):
        par_multi.gmp_nonlinear = [k, k, k]  # Three nonlinear orders of the GMP model; E.g., value = 1 --> Linear model
        par_multi.gmp_memory = [i+1, i+1, i+1]  # Three memory depths of the GMP model; E.g., value = 1 --> Memoryless
        par_multi.gmp_cross_len = [i+1, i+1]  # Two cross lengths of lag and lead envelopes, value; E.g., value = 1 --> cross length == 1
        
        # Number of parameters of the GMP model; Derivation shown in Section 3.2
        par_multi.num_par_gmp.append(k*(i+1) + 2*k*(i+1)*(i+1))
        
        # Number of multiplications for the part '|x(n-l)|^k', '|x(n-l-m)|^k', '|x(n-l-m)|^k' in GMP model, 
        # derivation shown in Section 3.2.2
        extra_multi = ((i+1) + 2*(i+1)**2)*(sum([(m+6)/2 for m in range(2, k+1, 2)]) + sum([(m+7)/2 for m in range(3, k+1, 2)]))
        
        # Total number of multiplications for the GMP model, derivation shown in Section 3.2.3
        par_multi.num_multi_gmp.append(3*par_multi.num_par_gmp[-1] + extra_multi)
        print('Number of parameters for GMP: ', par_multi.num_par_gmp[-1])
        print('Number of multiplications for GMP: ', par_multi.num_multi_gmp[-1])
        
        # Implemente ILA-GMP DPD method
        par_multi.dpd_methods(pa_in=par_multi.original_sig, pa_out=par_multi.y_power_align, method='ILA-GMP', 
                              pa_model=pa_model, matlab_eng=eng_multi)
        
    plt.figure(2)
    # Plot the NMSE results of ILA-GMP at current input memory depth
    plt.scatter(par_multi.num_multi_gmp[-k:], par_multi.nmse_ila_gmp[-k:], marker='s',
                c=color, label='M='+str(i)+', ILA-GMP')
    plt.figure(3)
    # Plot the ACPR results of ILA-GMP at current input memory depth
    plt.scatter(par_multi.num_multi_gmp[-k:], par_multi.acpr_ila_gmp[-k:], marker='s',
                c=color, label='M='+str(i)+', ILA-GMP')

plt.figure(2)
plt.xscale('log')
plt.xlim(left=1)
plt.xlabel('Number of multiplications')
plt.ylabel('NMSE (dB)')
plt.legend(loc=0)
plt.grid()
plt.figure(3)
plt.xscale('log')
plt.xlim(left=1)
plt.xlabel('Number of multiplications')
plt.ylabel('ACPR (dBc)')
plt.legend(loc=0)
plt.grid()
plt.show()
```

## Derivation of the number of multiplications in GMP

$$
\begin{aligned}
y_{\text{GMP}}(n) =& \sum_{k=0}^{K_a -1}\sum_{l=0}^{L_a -1}a_{kl}x(n-l)|x(n-l)|^k\\ 
&+ \sum_{k=0}^{K_b -1}\sum_{l=0}^{L_b -1}\sum_{m=0}^{M_b -1}b_{klm}x(n-l)|x(n-l-m)|^k\\
&+ \sum_{k=0}^{K_c -1}\sum_{l=0}^{L_c -1}\sum_{m=0}^{M_c -1}c_{klm}x(n-l)|x(n-l+m)|^k
\end{aligned}
$$

**Number of coefficients** $N_{\text{coef}} = K_a L_a + K_b L_b M_b + K_c L_c M_c$

**Number of multiplications** $N_{\text{multi}}$ = ?

### For the part $a_{kl}x(n-l)$, $b_{klm}x(n-l)$ and $c_{klm}x(n-l)$

Each element needs one complex multiplications. If we assume one complex multiplication counts for three real multiplications ([source](http://mathworld.wolfram.com/ComplexMultiplication.html)), the number of multiplications for theses parts would be:  
$$3(K_a L_a + K_b L_b M_b + K_c L_c M_c) = 3N_{\text{coef}}$$

###  For the part $|x(n-l)|^k$, $|x(n-l-m)|^k$ and $|x(n-l+m)|^k$:
#### if k is **even**:   
$|x(n-l)|^k = (|x(n-l)|^2)^{k/2} = \left (\Re(x(n-l))^2 + \Im(x(n-l))^2 \right)^{k/2}$   
where $\Re(x(n-l))^2 + \Im(x(n-l))^2$ accounts for **2** multiplications, $(\cdot)^{k/2}$ needs $(\frac{k}{2}-1)$ multiplications.   

**Proof:** For example, let's use $x$ to represent $\Re(x(n-l))^2 + \Im(x(n-l))^2$ in the below table  

k|Format|Multiplications, ($\frac{k}{2} -1$) 
:-:| :-:| :-:
2  |$x^1  \rightarrow x$ | 0
4  |$x^2  \rightarrow x\cdot x$ | 1
6  |$$x^3 \rightarrow x \cdot x \cdot x$$|2
8  |$x^4  \rightarrow x\cdot x \cdot x \cdot x$|3   

Although there are some fast computation algorithms for calculating integer exponentiation that require less number of multiplications, e.g., [Binary exponentiation](https://en.wikipedia.org/wiki/Exponentiation_by_squaring#cite_ref-frey_1-0), [Addition-chain exponentiation](https://en.wikipedia.org/wiki/Addition-chain_exponentiation), they perform better for very large exponents and need more memory for storing previous exponents which also consumes hardware resources. And because the highest nonlinear order $K$ used in GMP is usually **smaller than 10**, we consider $(\cdot)^{k/2}$ accounts for $\frac{k}{2} -1$ multiplications.   

Then, consider the multiplication between $|x(n-l)|^k$ and the complex-valued $x(n-l)$ as **2** multiplications, the number of multiplications $N_{\text{multi}}$ in terms of each $l$ is $\frac{k}{2} -1 +2 +2 = \frac{k}{2} +3$. So the number of multiplications for even $k$ 

$$
N_{\text{multi}} = L\cdot \sum_{k=2,k \text{ is even}}^{K}(\frac{k}{2} +3)
$$

#### if k is **odd**:   
$|x(n-l)|^k = (|x(n-l)|^2)^{\frac{(k-1)}{2}}|x(n-l)|^{\frac{1}{2}} = \left (\Re(x(n-l))^2 + \Im(x(n-l))^2 \right)^{\frac{(k-1)}{2}}|x(n-l)|^{\frac{1}{2}}$   
Similarly, the part $\left (\Re(x(n-l))^2 + \Im(x(n-l))^2 \right)^{\frac{(k-1)}{2}}$ needs $\frac{k-1}{2}-1+2$ multiplications before being multiplied by $\sqrt{|x(n-l)|}$. And if we assume the $\sqrt{\cdot}$ function could be approximated using a LUT, the number of multiplications $N_{\text{multi}}$ in terms of each $l$ is $\frac{k-1}{2} -1 +2 +1 +2$. So the multiplications for odd $k$ would be:

$$
N_{\text{multi}} = L\cdot \sum_{k=3,k \text{ is odd}}^{K}(\frac{k-1}{2} +4)
$$

### In total for MP, lagging and leading envelopes
So the total number of multiplications for the **MP envelope** $\sum_{k=0}^{K_a -1}\sum_{l=0}^{L_a -1}a_{kl}x(n-l)|x(n-l)|^k$ is:

$$
N_{\text{multi-MP}} = 3 K_a L_a + L_a\cdot \left(\sum_{k=2,k \text{ is even}}^{K}(\frac{k}{2} +3) + \sum_{k=3,k \text{ is odd}}^{K}(\frac{k-1}{2} +4) \right)
$$

For the **lagging envelope**:
$$
N_{\text{multi-lag}} = 3 K_b L_b M_b +  L_b M_b\cdot \left(\sum_{k=2,k \text{ is even}}^{K}(\frac{k}{2} +3) + \sum_{k=3,k \text{ is odd}}^{K}(\frac{k-1}{2} +4) \right)
$$

For the **leading envelope**:
$$
N_{\text{multi-lead}} = 3 K_c L_c M_c +  L_c M_c\cdot \left(\sum_{k=2,k \text{ is even}}^{K}(\frac{k}{2} +3) + \sum_{k=3,k \text{ is odd}}^{K}(\frac{k-1}{2} +4) \right)
$$
**In total**, the GMP needs:
$$
\begin{aligned}
N_{\text{multi}} &= N_{\text{multi-MP}} + N_{\text{multi-lag}} + N_{\text{multi-lead}} \\
&= 3N_{\text{coef}}+ (L_a + L_b M_b + L_c M_c)\cdot \left(\sum_{k=2,k \text{ is even}}^{K}(\frac{k}{2} +3) + \sum_{k=3,k \text{ is odd}}^{K}(\frac{k-1}{2} +4) \right)
\end{aligned}
$$
